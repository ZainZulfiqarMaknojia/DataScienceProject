### Data Preperation

```python
df.apply(lambda x:sum(x.isnull()),axis=0)
```

```python
df['Gender'].fillna('Male',inplace=True)
df['Married'].fillna('Yes',inplace=True)
df['Dependents'].fillna(0,inplace=True)
df['Self_Employed'].fillna('No',inplace=True)
df['Credit_History'].fillna(1,inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
df['Loan_Amount_Term'].fillna(360,inplace=True)
```

```python
df.apply(lambda x:sum(x.isnull()),axis=0)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
df.Dependents = df.Dependents.astype('str')
featureVariable=['Gender','Married','Dependents',
                 'Education','Self_Employed',
                 'Property_Area','Loan_Status']
L=LabelEncoder()
for i in featureVariable:
    df[i]=L.fit_transform(df[i])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['totalIncome'] = df.ApplicantIncome + df.CoapplicantIncome
```

```python
df.head()
```

```python
Loan_ID = df.iloc[:,:1]
X = df.drop(['ApplicantIncome','CoapplicantIncome','Loan_ID'],axis = 1)
X
```

```python
Loan_ID
```

```python
y = X.iloc[:,9:10]
X = X.drop('Loan_Status',axis=1)
y
```

```python
X
```

# Data Modeling And Importing Necessary Libraries

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=21)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
LR=LogisticRegression(solver='liblinear')
LR.fit(X_train,y_train.values.ravel())
Prediction1=LR.predict(X_test)
Accuracy1=metrics.accuracy_score(Prediction1,y_test)
print("Logistic Regression: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN3=KNeighborsClassifier(n_neighbors=3)
Features=['Credit_History','Gender','Married','Education']
KNN3.fit(X_train[Features],y_train.values.ravel())
Prediction2=KNN3.predict(X_test[Features])
Accuracy2=metrics.accuracy_score(Prediction2,y_test)
print("KNN Model with 3 neighbours: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy2))
```

```python
KNN5=KNeighborsClassifier(n_neighbors=5)
Features=['Credit_History','Gender','Married','Education']
KNN5.fit(X_train[Features],y_train.values.ravel())
Prediction3=KNN5.predict(X_test[Features])
Accuracy3=metrics.accuracy_score(Prediction3,y_test)
print("KNN Model with 5 neighbours: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy3))
```

```python
KNN10=KNeighborsClassifier(n_neighbors=10)
Features=['Credit_History','Gender','Married','Education']
KNN10.fit(X_train[Features],y_train.values.ravel())
Prediction4=KNN10.predict(X_test[Features])
Accuracy4=metrics.accuracy_score(Prediction4,y_test)
print("KNN Model with 10 neighbours: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy4))
```

```python

from sklearn.tree import DecisionTreeClassifier

DTC=DecisionTreeClassifier()
Features=['Credit_History','Gender','Married','Education']
DTC.fit(X_train[Features],y_train.values.ravel())
Prediction5=DTC.predict(X_test[Features])
Accuracy5=metrics.accuracy_score(Prediction5,y_test)
print("Decision Tree: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy5))
```

```python
from sklearn.ensemble import BaggingClassifier
BCK= BaggingClassifier(KNeighborsClassifier(),max_samples=0.50, max_features=0.5)
Features=['Education','Loan_Amount_Term','Credit_History','Dependents','Property_Area']
BCK.fit(X_train[Features],y_train.values.ravel())
Prediction6=BCK.predict(X_test[Features])
Accuracy6=metrics.accuracy_score(Prediction6,y_test)
print("Bagging using KNN Model: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy6))
```

```python
BCD = BaggingClassifier(DecisionTreeClassifier(),max_samples=0.50, max_features=0.5)
Features=['Education','Loan_Amount_Term','Credit_History','Dependents','Property_Area']
BCD.fit(X_train[Features],y_train.values.ravel())
Prediction7=BCD.predict(X_test[Features])
Accuracy7=metrics.accuracy_score(Prediction7,y_test)
print("Bagging using Decision Tree: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy7))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(n_estimators=50,min_samples_split=25,max_depth=7,max_features=1)
RFC.fit(X_train,y_train.values.ravel())
Prediction8=RFC.predict(X_test)
Accuracy8=metrics.accuracy_score(Prediction8,y_test)
print("Random Forest: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy8))
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
Features=['Credit_History','Gender','Married','Education']
GNB.fit(X_train[Features],y_train.values.ravel())
Prediction9=GNB.predict(X_test[Features])
Accuracy9=metrics.accuracy_score(Prediction9,y_test)
print("Naive Baise: ")
print("Accuracy : %s" % "{0:.5%}\n".format(Accuracy9))
```
